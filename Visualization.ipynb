```python
import pandas as pd # грузим pandas
import matplotlib.pyplot as plt # грузим matplotlib
```

```python
df = pd.read_csv('booking_data_homework.csv') # загружаем наш набор данных в pandas
df
```

```python
# группируем наш набор данных по destination_region и purpose_trip, затем считаем количество поездок
data = df.groupby(['destination_region', 'purpose_trip'])['purpose_trip'].count().unstack('purpose_trip')
data
```

```python
data.plot(kind='bar') # строим простой график в одну строчку
```

```python
# строим простой график в одну строчку, но используем классный параметр stacked, чтобы столбки были друг над другом
data.plot(kind='bar', stacked=True) 
```

```python
# строим более гибкий график с кучей опций

width = 0.6 # ширина столбика

plt.rcParams["figure.figsize"] = (20,10) # размер картинки
# первая группа столбиков (их названия, значения, ширина, цвет)
plt.bar(data.index, data['business'], width, color='#d62728')
# вторая группа столбиков, bottom означает что она стоит над первой
plt.bar(data.index, data['leasure'], width, color='#3498eb',  bottom=data['business'])
plt.xlabel('Destinations', size=15) # подпись оси x и размер шрифта
plt.ylabel('Orders', size=15) # подпись оси y и размер шрифта
plt.title('Bookings', size=25) # подпись графика и размер шрифта
plt.xticks(data.index) # сами подписи
plt.legend(('Business', 'Leasure')) # легенда
plt.show() # вывод графика
```
